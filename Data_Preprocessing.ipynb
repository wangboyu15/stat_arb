```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white", palette="muted", color_codes=True)
from sklearn.metrics import mean_squared_error
```

```python
data_path = f"data"
data = pd.read_csv(f"{data_path}/data.csv")
data
```

```python
date = data.groupby(f"date")["ticker"].count()
date
```

```python
price = data.pivot(index='date', columns='ticker', values='last')
price
```

```python
price.to_csv(f"{data_path}/price.csv")
```

```python
rtn = price.pct_change()
rtn
```

```python
rtn.to_csv(f"{data_path}/rtn.csv")
```

```python
volume = data.pivot(index='date', columns='ticker', values='volume')
volume
```

```python
volume.to_csv(f"{data_path}/volume.csv")
```
