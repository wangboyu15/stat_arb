```python
class Factor:
    def __init__(self, factor_name: str, reverse:bool=False, data_path: str="data", factor_path: str="factors"):
        self.factor = pd.DataFrame()
        self.factor_name = factor_name
        self.reverse = reverse
        self.data_path = data_path
        self.factor_path = factor_path
    
    def cal_factor(self, *args, **kwargs):
        pass
    
    def handle_data(self):
        if self.factor.empty: return
        self.factor = self.factor.replace([np.inf, -np.inf], np.nan)
        
        '''shift(1), otherwise look-head bias!'''
        self.factor = self.factor.shift(1)
        
        if self.reverse:
            self.factor *= -1.0
    
    def save_factor(self):
        self.handle_data()
        if not self.factor.empty:
            self.factor.to_csv(f"{self.factor_path}/{self.factor_name}.csv")
            display(self.factor)
    
    def run(self):
        self.cal_factor()
        self.save_factor()
        
```

```python
class Reversal(Factor):
    
    def __init__(self, factor_name: str, window_len: int, reverse:bool=False, data_path: str="data", factor_path: str="factors"):
        super().__init__(factor_name, reverse, data_path, factor_path)
        self.window_len = window_len
    
    def cal_factor(self):
        rtn = pd.read_csv(f"{self.data_path}/rtn.csv", index_col = 0)
        self.factor = -1.0 * rtn.rolling(self.window_len).mean()
        
```

```python
class HistoricalMoments(Factor):
    
    def __init__(self, factor_name: str, window_len: int, moments: int, reverse: bool=False, data_path: str="data", factor_path: str="factors"):
        super().__init__(factor_name, reverse, data_path, factor_path)
        self.window_len = window_len
        self.moments = moments
        self.reverse = reverse
    
    def cal_factor(self):
        rtn = pd.read_csv(f"{self.data_path}/rtn.csv", index_col = 0)
        
        if self.moments == 2:
            self.factor = rtn.rolling(self.window_len).std()
        elif self.moments == 3:
            self.factor = rtn.rolling(self.window_len).skew()
        elif self.moments == 4:
            self.factor = rtn.rolling(self.window_len).kurt()
        else:
            raise NotImplementedError("This moment is not yet implemented")
        
```

```python
class Illiquidity(Factor):
    
    def __init__(self, factor_name: str, window_len: int, reverse: bool=False, data_path: str="data", factor_path: str="factors"):
        super().__init__(factor_name, reverse, data_path, factor_path)
        self.window_len = window_len
        self.reverse = reverse
    
    def cal_factor(self):
        rtn = pd.read_csv(f"{self.data_path}/rtn.csv", index_col = 0)
        volume = pd.read_csv(f"{self.data_path}/volume.csv", index_col = 0)
        
        self.factor = (1 - (rtn.abs() / volume).rolling(self.window_len).mean()).applymap(np.log)
```

```python
class Salience(Factor):
    
    def __init__(self, factor_name: str, reverse: bool=False, data_path: str="data", factor_path: str="factors"):
        super().__init__(factor_name, reverse, data_path, factor_path)
        self.reverse = reverse
        self.epsilon = 0.1
    
    def cal_factor(self):
        rtn = pd.read_csv(f"{self.data_path}/rtn.csv", index_col = 0)
        cross_sec_mean = rtn.mean(axis=1)        
        self.factor = (rtn.sub(cross_sec_mean, axis='index')) / (rtn.abs().add(cross_sec_mean.abs(), axis='index') + self.epsilon)
```

```python
salience = Salience(factor_name="salience", reverse=True)
salience.run()
```

```python
class RankCrossSection(Factor):
    
    def __init__(self, factor_name: str, window_len: int, reverse: bool=False, data_path: str="data", factor_path: str="factors"):
        super().__init__(factor_name, reverse, data_path, factor_path)
        self.window_len = window_len
        self.reverse = reverse
    
    def cal_factor(self):
        rtn = pd.read_csv(f"{self.data_path}/rtn.csv", index_col = 0)
        rank = rtn.rank(ascending=True, axis = 1, pct = True)
        self.factor = rank.rolling(self.window_len).mean()
```

```python
class Jump(Factor):
    
    def __init__(self, factor_name: str, window_len: int, reverse: bool=False, data_path: str="data", factor_path: str="factors"):
        super().__init__(factor_name, reverse, data_path, factor_path)
        self.window_len = window_len
        self.reverse = reverse
    
    def cal_factor(self):
        rtn = pd.read_csv(f"{self.data_path}/rtn.csv", index_col = 0)
        price = pd.read_csv(f"{self.data_path}/price.csv", index_col = 0)
        log_price = price.applymap(np.log)
        log_rtn = log_price - log_price.shift(1)
        
        mat = 2 * (rtn - log_rtn) - log_rtn ** 2
        
        self.factor = mat.rolling(self.window_len).mean()
        
```

```python
jump = Jump(factor_name=f"jump_5d", window_len=5, reverse=True)
jump.run()
```

```python
jump = Jump(factor_name=f"jump_22d", window_len=22, reverse=True)
jump.run()
```

```python
class Bollinger(Factor):
    
    def __init__(self, factor_name: str, window_len: int, reverse: bool=False, data_path: str="data", factor_path: str="factors"):
        super().__init__(factor_name, reverse, data_path, factor_path)
        self.window_len = window_len
        self.reverse = reverse
        self.c = 2
    
    def cal_factor(self):
        rtn = pd.read_csv(f"{self.data_path}/rtn.csv", index_col = 0)
        vol = rtn.rolling(self.window_len).std()
        
        self.factor = rtn + self.c * vol
        
```

```python
bollinger = Bollinger(factor_name=f"bollinger_5d", window_len=5, reverse=True)
bollinger.run()
```

```python
bollinger = Bollinger(factor_name=f"bollinger_22d", window_len=22, reverse=True)
bollinger.run()
```
